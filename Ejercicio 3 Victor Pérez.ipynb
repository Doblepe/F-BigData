```python
# Datos para el gráfico Pie Chart
labels = ['Manzanas', 'Naranjas', 'Bananas', 'Mangos']
sizes = [25, 30, 20, 25]

# Crea el gráfico de "pie chart"
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Ajusta el aspecto del gráfico
plt.axis('equal')

# Muestra el gráfico
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Datos para el gráfico 3D
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [0, 0, 0, 0, 0]

# Crea una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grafica los datos
ax.scatter(x, y, z)

# Establece las etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Muestra el gráfico
plt.show()
```

```python
# Los “diagramas de cajas” son un elemento importante de los mismos. Encuentra un ejemplo donde poder mostrar estos gráficos con Python simplemente se pide mostrar esas Gráficos con el ejemplo.
import numpy as np

# Crea el conjunto de datos
data = np.random.normal(0, 1, 100)

# Crea el diagrama de cajas
fig, ax = plt.subplots()
ax.boxplot(data)

# Ajusta el aspecto del gráfico
ax.set_xlabel('Datos')
ax.set_ylabel('Valores')
ax.set_title('Diagrama de Cajas')

# Muestra el gráfico
plt.show()
```

```python
# ¿ Qué son Q1, Q2, Q3 ? 
```

```python
# En estadística, Q1, Q2 y Q3 se refieren a los cuartiles de un conjunto de datos, es decir, los valores que dividen al conjunto de datos en cuatro partes iguales. Q2 se refiere al segundo cuartil, que también se conoce como la mediana del conjunto de datos. Es el valor que divide al conjunto de datos en dos partes iguales. La mediana es útil para resumir la posición central de un conjunto de datos.
```

```python
# Son valores atípicos en español,es decir se encuentran significativamente alejados de los valores típicos en un conjunto de datos. En los diagramas de cajas, los outliers se representan como puntos individuales más allá de los bigotes que se extienden desde la caja del diagrama.  
```
