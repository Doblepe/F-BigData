```python
# Importamos las librer√≠as necesarias
from plotnine import *
import pandas as pd
df = pd.read_csv('iris.csv')

# Cargamos los datos de Iris Dataset en un DataFrame de pandas
iris_df = pd.DataFrame(data=df)

# Creamos una gr√°fica de dispersi√≥n
ggplot(iris_df, aes(x='sepal.length', y='sepal.width', color='variety')) + \
geom_point() + \
labs(title='Gr√°fica de dispersi√≥n de Iris Dataset', x='Longitud del s√©palo', y='Anchura del s√©palo')
```
